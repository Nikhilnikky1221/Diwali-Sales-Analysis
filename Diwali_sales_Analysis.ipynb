```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")       #unicode parameter is passed for reading special characters
print(df.shape)     #shape is attribute so directly we can call without ()
```

```python
print(df.head(10))
```

```python
df.info()
```

# Data Cleaning


```python
# drop Status and unnamed1 columns              
# axis parameter implies deleting the whole column
# inplace parameter implies that the new data frame will be used inplace of old  

df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
print(df.head(10))
```

```python
# checking for columns with null values
print(df.isnull().sum())

# removing rows containing null values
df.dropna(inplace=True)     
print(df.shape)
```

```python
# changing data types of columns
df['Amount']=df['Amount'].astype('int')
df.info()
```

```python
# columns is an attribute of data types
print(df.columns)
```

```python
# Renaming columns
df.rename(columns={'Marital_Status':"shaadi"})
```

```python
# statistical information of columns with int or float data types
print(df.describe())
print(df['Age'].describe())
print(df[['Age','Orders','Amount']].describe())
```

# Exploratory Data Analysis


By Gender

```python
# no of purchases done vs gender
p1=sns.countplot(x="Gender",data=df)
for p in p1.containers:
    p1.bar_label(p)
plt.title("Gender VS Count")
plt.show()
```

```python
# purchase Amount vs gender
gender_count=df.groupby(["Gender"], as_index=False)[["Orders","Amount"]].sum().sort_values(by="Orders",ascending=False)
# print(gender_count)

#df["Gender"][df["Gender"]=="F"].count()   will give us count if condition is true
count=[df["Gender"][df["Gender"]=="F"].count(),df["Gender"][df["Gender"]=="M"].count()]
gender_count["count"]= count

gender_count["Avg Amount per person"]=[gender_count['Amount'][0]/gender_count['count'][0],gender_count['Amount'][1]/gender_count['count'][1]]

gender_count["Avg Amount per order"]=[gender_count['Amount'][0]/gender_count['Orders'][0],gender_count['Amount'][1]/gender_count['Orders'][1]]
print(gender_count)

p1=sns.barplot(x="Gender",y="Amount",data=gender_count)
for p in p1.containers:
    p1.bar_label(p)
plt.title("Gender VS Count")
plt.show()
```

By Age Group


```python
plt.hist(df["Age"],density=True,bins=20)
plt.title("Age distribution")
sns.kdeplot(df["Age"],bw_adjust=1)
plt.show()
```

```python
k=df.groupby(["Age Group"],as_index=True)["Age Group"].count()
print(type(k))
print(k)
print(k.index)
plt.bar(k.index,k)
print(k[0],k[1],k[2],k.index[3])
plt.show()
```

```python
# countplot of Age groups
p1=sns.countplot(data=df, x="Age Group",order=['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'])
for p in p1.containers:
    p1.bar_label(p)
plt.show()

# countplot of Age groups diff
p2=sns.countplot(data=df, x="Age Group",hue="Gender",order=['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'])
for p in p2.containers:
    p2.bar_label(p)
plt.show()

# stacked bar plot
p3=sns.countplot(data=df,x="Age Group",hue="Gender",order=['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'], dodge=False)
for p in p3.containers:
    p3.bar_label(p)
plt.show()
```

```python
age_group_count=df.groupby(["Age Group"],as_index=False)[["Orders","Amount"]].sum()
age_group_count['Amount in Million']=age_group_count["Amount"]/1000000
print(age_group_count)
p5=sns.barplot(x="Age Group",y="Amount in Million",data=age_group_count)
for p in p5.containers:
    p5.bar_label(p)

plt.show()
```

```python
age_group_count=df.groupby(["Age Group","Gender"],as_index=False)[["Orders","Amount"]].sum()
age_group_count['Amount in Million']=age_group_count["Amount"]/1000000
print(age_group_count)
plt.figure(figsize=(8,7))
p5=sns.barplot(x="Age Group",y="Amount in Million",data=age_group_count, hue="Gender",dodge=False)
for p in p5.containers:
    p5.bar_label(p)
plt.show()
```

By State


```python
sns.countplot(data=df, x="Zone")
plt.show()

Zone_count=df.groupby(["Zone"],as_index=False)[["Orders","Amount"]].sum().sort_values("Orders",ascending=False)
p6=sns.barplot(x="Zone",y="Orders",data=Zone_count)
for p in p6.containers:
    p6.bar_label(p)
plt.title("Zone VS Orders")
plt.show()

Zone_count=Zone_count.sort_values("Amount",ascending=False)
p7=sns.barplot(x="Zone",y="Amount",data=Zone_count)
for p in p7.containers:
    p7.bar_label(p)
plt.title("Zone VS Amount")
plt.show()
```

```python
State_count=df.groupby(["State"],as_index=False)[["Orders","Amount"]].sum().sort_values("Orders",ascending=False)
plt.figure(figsize=(25,8))
p6=sns.barplot(x="State",y="Orders",data=State_count)
for p in p6.containers:
    p6.bar_label(p)
plt.title("State VS Orders")
plt.show()

State_count=State_count.sort_values("Amount",ascending=False)
plt.figure(figsize=(25,8))
p7=sns.barplot(x="State",y="Amount",data=State_count)
for p in p7.containers:
    p7.bar_label(p)
plt.title("State VS Amount")
plt.show()
```
